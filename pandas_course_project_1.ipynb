#### <u>***Q1: TODO - Which are the major categories with the TOP 10 graduate salaries ?*** </u>

Step 1:
- Grouping data by major category
- Finding mean of graduate median salaries
- Sort values in descending order

```python
top_major_cat = grad_df.groupby('major_cat').agg({'grad_median':'mean'})
top_major_cat.sort_values(by='grad_median',ascending=False,inplace=True)
```

Step 2 - Plotting:
- 'BARH' plots a horizontal barchart with numeric values for each category

```python
fig,ax = plt.subplots()
bars = ax.barh(top_major_cat[:10].index,top_major_cat[:10].grad_median);
plt.xticks(np.arange(0,95000,10000));
ax.bar_label(bars);
plt.title('Top 10 Graduate Salaries');
```

Answers & Observations:
- Engineering has the highest median salary of 94k 
- Physical Sciences, Comp & Math, Business are also pretty high up in the 84k+ range
- This is followed by social science and law in the 70k range
- Humanities, Pyschology, Communications make around 65k which is a good salary

#### <u>***Q3: TODO - In Health ( Major Category ) what is the share of nongraduates to graduates employed for top 3 majors with highest overall employment ?*** </u>

Steps:
- Creating a new column summing non-graduate & graduate employed total
- Creating a daframe with all entries in Health category
- Sort it by total_employed in descending order
- Plot the top 3 rows

A piechart is a graph in which a circle is divided into sectors to represent proportions / percentages of data 


```python
grad_df['total_employed'] = grad_df['grad_employed'] + grad_df['nongrad_employed']

health_df = grad_df[grad_df.major_cat == 'Health'][['major','grad_employed','nongrad_employed','total_employed']]
health_df.sort_values(by='total_employed',inplace=True,ascending=False)

my_label = ['Grad Employed','Nongrad Employed']
my_color = ['cadetblue','powderblue']

plt.title(f" Major : {health_df.iloc[0].major}");
plt.pie(health_df.iloc[0][['grad_employed','nongrad_employed']],labels=my_label,autopct='%1.1f%%',colors=my_color,shadow=True,wedgeprops = {'linewidth': 3});
```

```python
plt.title(f" Major : {health_df.iloc[1].major}");
plt.pie(health_df.iloc[1][['grad_employed','nongrad_employed']],labels=my_label,autopct='%1.1f%%',colors=my_color,shadow=True,wedgeprops = {'linewidth': 3});
```

```python
plt.title(f" Major : {health_df.iloc[2].major}");
plt.pie(health_df.iloc[2][['grad_employed','nongrad_employed']],labels=my_label,autopct='%1.1f%%',colors=my_color,shadow=True,wedgeprops = {'linewidth': 3});
```

Answers & Observations:
